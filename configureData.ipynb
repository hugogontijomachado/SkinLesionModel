```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import cv2
import random
```

```python
DATA_PATH = '/home/ec2-user/isic-data'
TRAIN_PATH = '/home/ec2-user/isic-data/train'
DUP_PATH = '/home/ec2-user/isic-data/duplicates'

df = pd.read_csv(os.path.join(DATA_PATH,'ISIC_2020_Training_GroundTruth.csv'))
df_v2 = pd.read_csv(os.path.join(DATA_PATH,'ISIC_2020_Training_GroundTruth_v2.csv'))
df_dup = pd.read_csv(os.path.join(DATA_PATH,'ISIC_2020_Training_Duplicates.csv'))
```

```python
### Esta função irá excluir as entradas de imagens repetidas dos dataframes df e df_v2
### Além disso irá mover estas imagens da pasta TRAIN_PATH para a pasta DUP_PATH caso elas não tenham sido movidas
def exclude_duplicates():
    for im in list(df_dup['image_name_2']):
        df.drop( df.index[df['image_name'] == im], inplace = True )
        df.reset_index(drop=True, inplace=True)
    
        df_v2.drop( df_v2.index[df_v2['image_name'] == im], inplace = True )
        df_v2.reset_index(drop=True, inplace=True)
    
        filename = os.path.join(TRAIN_PATH, "{}.jpg".format(im))
    
        if os.path.isfile(filename):
            newfilename = os.path.join(DUP_PATH, "{}.jpg".format(im) )
            os.replace(filename, newfilename)

exclude_duplicates()
```

```python
### Esta função irá mostrar as 2 imagens duplicadas da linha N do df_dup
def show_duplicate(n):
    print(n)
    filename1 = os.path.join(TRAIN_PATH, "{}.jpg".format( df_dup['image_name_1'][n] ) )
    if not os.path.isfile(filename1): filename1 =  os.path.join(DUP_PATH, "{}.jpg".format( df_dup['image_name_1'][n] ) )
    
    filename2 = os.path.join(TRAIN_PATH, "{}.jpg".format( df_dup['image_name_2'][n] ) )
    if not os.path.isfile(filename2): filename2 = os.path.join(DUP_PATH, "{}.jpg".format( df_dup['image_name_2'][n] ) )
    
    img1 = cv2.imread( filename1 )
    img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

    img2 = cv2.imread( filename2 )
    img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

    fig, ax = plt.subplots(1,2)
    ax[0].imshow(img1)
    ax[1].imshow(img2)

    
n = random.randint(0,len(df_dup)-1)
show_duplicate(n)
```

```python
### Esta função irá testar se todas as imagens do dataframe que foi passado estão dentro de TRAIN_PATH
### Da mesma maneira, irá testar se todas as imagens dentro de TRAIN_PATH estão no dataframe em questão
def test_integrity(df):
    files = os.listdir(TRAIN_PATH)
    for name in df['image_name']:
        if "{}.jpg".format(name) not in files:
            print(name)

    names = list(df['image_name'])
    for file in files:
        if file.replace('.jpg','') not in names:
            print(file)

test_integrity(df)
test_integrity(df_v2)
```
